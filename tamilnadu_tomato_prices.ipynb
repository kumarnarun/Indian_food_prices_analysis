The dataset consists of prices of tomatoes at various time periods and also at different markets in Tamilnadu.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tn_tomato_price.csv')
df.head()
```

```python
df['date'].unique()
```

## This is the average price of the tomatoes in different markets of tamilnadu in past 10 years

```python
df.groupby('market').mean().reset_index().sort_values(by='price',ascending = False)
```

```python
df.groupby('date').mean().reset_index().sort_values(by='price',ascending = False)
```
