```python
import itertools
import pandas as pd
```

create 5 list of some values
for example config = [RAM, hard] like 5 list
all the combinations of rows
final column result

```python
friends = ['arun','parthi','amsa']
cricket = ['virat','dhoni','rohit']
tennis = ['federer','nadal','alcaraz']
chess = ['magnus','nepo','anand']
shuttle = ['sindhu','newal','lakshaya']
```

```python
combo = []
for x in friends:
    for y in cricket:
        for z in tennis:
            for c in chess:
                for s in shuttle:
                    l = [x,y,z,c,s]
                    combo.append(l)
pd.DataFrame(combo,columns = ['friends','cricket','tennis','chess','shuttle'])          
```

```python
a = list(itertools.combinations([friends,cricket,tennis,chess,shuttle],5))
```

```python
for i in a:
    print(i)
```
